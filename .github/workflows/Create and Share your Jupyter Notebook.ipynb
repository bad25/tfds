```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pyplot
import scipy.stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
import numpy as np
from sklearn.linear_model import LinearRegression
```

#### Read the dataset in the csv file from the URL


```python
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
resp = await fetch(URL)
boston_url = io.BytesIO((await resp.arrayBuffer()).to_py())
```

```python
boston_df=pd.read_csv(boston_url)
```

```python
boston_df.head()
```

```python
# Create a boxplot for the median value of owner-occupied homes (MEDV)
pyplot.figure(figsize=(8, 6))
pyplot.boxplot(boston_df['MEDV'], vert=False)
pyplot.xlabel('Median Value of Owner-Occupied Homes (MEDV)')
pyplot.title('Boxplot of MEDV')
pyplot.show()
```

```python
# Count the occurrences of each unique value in the 'CHAS' column
chas_counts = boston_df['CHAS'].value_counts()
# Create a bar plot for the 'CHAS' variable
pyplot.figure(figsize=(6, 4))
chas_counts.plot(kind='bar', color='skyblue')
pyplot.xlabel('Charles River (CHAS)')
pyplot.ylabel('Count')
pyplot.title('Bar Plot of CHAS')
pyplot.xticks(rotation=0, ticks=[0, 1], labels=['Not on Charles River', 'On Charles River'])
pyplot.show()
```

```python
# Create a boxplot for the MEDV variable vs the AGE variable
pyplot.figure(figsize=(10, 6))
pyplot.boxplot([boston_df[boston_df['AGE'] < 50]['MEDV'], 
             boston_df[(boston_df['AGE'] >= 50) & (boston_df['AGE'] < 75)]['MEDV'],
             boston_df[boston_df['AGE'] >= 75]['MEDV']],
            labels=['AGE < 50', '50 <= AGE < 75', 'AGE >= 75'])
pyplot.xlabel('Age Categories')
pyplot.ylabel('Median Value of Owner-Occupied Homes (MEDV)')
pyplot.title('Boxplot of MEDV vs AGE')
pyplot.show()
```

```python
# Create a scatter plot for NOX vs INDUS
pyplot.figure(figsize=(10, 6))
pyplot.scatter(boston_df['INDUS'], boston_df['NOX'], color='skyblue')
pyplot.xlabel('Proportion of Non-Retail Business Acres per Town (INDUS)')
pyplot.ylabel('Nitric Oxide Concentrations (NOX)')
pyplot.title('Scatter Plot: NOX vs INDUS')
pyplot.grid(True)
pyplot.show()
```

```python
# Create a histogram for the PTRATIO variable
pyplot.figure(figsize=(8, 6))
pyplot.hist(boston_df['PTRATIO'], bins=15, color='skyblue')
pyplot.xlabel('Pupil-Teacher Ratio (PTRATIO)')
pyplot.ylabel('Frequency')
pyplot.title('Histogram of PTRATIO')
pyplot.grid(True)
pyplot.show()
```

```python
# Extract the features (DIS) and target (MEDV)
X = boston_df[['DIS']]
y = boston_df['MEDV']

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Coefficient for DIS
coefficient = model.coef_[0]
print('Coefficient for DIS:', coefficient)
```

```python
# Create a histogram for the PTRATIO variable
pyplot.figure(figsize=(8, 6))
pyplot.hist(boston_df['PTRATIO'], bins=15, color='skyblue')
pyplot.xlabel('Pupil-Teacher Ratio (PTRATIO)')
pyplot.ylabel('Frequency')
pyplot.title('Histogram of Pupil-Teacher Ratio (PTRATIO)')
pyplot.grid(True)
pyplot.show()
```

```python
# Create a scatter plot for NOX vs INDUS
pyplot.figure(figsize=(10, 6))
pyplot.scatter(boston_df['INDUS'], boston_df['NOX'], color='skyblue')
pyplot.xlabel('Proportion of Non-Retail Business Acres per Town (INDUS)')
pyplot.ylabel('Nitric Oxide Concentrations (NOX)')
pyplot.title('Scatter Plot: Nitric Oxide Concentrations vs. INDUS')
pyplot.grid(True)
pyplot.show()
```
